```python
import pandas as pd
import numpy as np
from IPython.display import display
from google.colab import files
```

```python
ball_by_ball = pd.read_csv('BallByBall.csv', low_memory=False)
performance = pd.read_csv('team_performance.csv')
```

```python
display(ball_by_ball.head())
display(performance.head())
```

```python
print("Ball By Ball Data Info:")
ball_by_ball.info()
print("\nPerformance Info:")
performance.info()
```

```python
print("\nMissing values in Ball By Ball before filling:")
print(ball_by_ball.isnull().sum())
```

```python
ball_by_ball['type of extras'] = ball_by_ball['type of extras'].fillna('None')
ball_by_ball['wicket_type'] = ball_by_ball['wicket_type'].fillna('Not Out')
ball_by_ball['fielders_involved'] = ball_by_ball['fielders_involved'].fillna('NA')
ball_by_ball['Player Out'] = ball_by_ball['Player Out'].fillna('NA')
```

```python
ball_by_ball['Season'] = pd.to_numeric(ball_by_ball['Season'], errors='coerce')
ball_by_ball['Date'] = pd.to_datetime(ball_by_ball['Date'], errors='coerce')
```

```python
ball_by_ball.columns = ball_by_ball.columns.str.replace(' ', '_')
```

```python
performance['Second_Innings_Score'] = performance['Second_Innings_Score'].fillna(0)
performance['Win_Type'] = performance['Win_Type'].fillna('NA')
performance['Win_Margin'] = performance['Win_Margin'].fillna(0)
performance['Player_of_Match'] = performance['Player_of_Match'].fillna('NA')
```

```python
performance.columns = performance.columns.str.replace(' ', '_')
```

```python
ball_by_ball.to_csv('cleaned_BallByBall.csv', index=False)
performance.to_csv('cleaned_Team_Performance.csv', index=False)
```

```python
print("Cleaned files saved: cleaned_BallByBall.csv & cleaned_Team_Performance.csv")
```
